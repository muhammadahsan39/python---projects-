```
# Converting date column into datetime format

df['date'] = pd.to_datetime(df['date'])
```

```
df["date"]
```

```
df['Year'] = df['date'].dt.year
df['Month'] = df['date'].dt.month
df['Day'] = df['date'].dt.day
```

```
# aqi distribution graph
```

```
# AQI Distribution

plt.figure()
sns.histplot(df['aqi_pm2.5'], bins=30, kde=True)

plt.title("Distribution of AQI Levels in Lahore")
plt.xlabel("AQI")
plt.ylabel("Frequency")

plt.show()
```

```
# AQI Trend Analysis

plt.figure()

plt.plot(df['date'], df['aqi_pm2.5'])

plt.title("AQI Trend Over Time")
plt.xlabel("Date")
plt.ylabel("AQI")

plt.xticks(rotation=45)

plt.show()
```

```
#boxplot (outliar)
```

```
# Detecting Outliers in AQI

plt.figure()

sns.boxplot(x=df['aqi_pm2.5'])

plt.title("AQI Outlier Detection")

plt.show()
```

```
#correlation heat map
```

```
# Correlation Heatmap

plt.figure(figsize=(8,6))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Between Numerical Features")

plt.show()
```

```
# monthly aqi analysis
```

```
# Monthly Average AQI

monthly_aqi = df.groupby('Month')['aqi_pm2.5'].mean()

monthly_aqi
```

```
# Monthly AQI Visualization

monthly_aqi.plot(kind='bar')

plt.title("Average Monthly AQI")
plt.xlabel("Month")
plt.ylabel("Average AQI")

plt.show()
```
